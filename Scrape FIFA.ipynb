# Approach 2
Scrape the homepage through 663 pages

```python
dataframe = pd.DataFrame(columns=dataframeColumns())
for pageNum in range(0,664):
    page = requests.get("https://www.easports.com/fifa/ultimate-team/api/fut/item?page={}".format(pageNum))
    addPlayersFromPage(page.json(), dataframe)
    print '{0} of {1}\r'.format(pageNum, 663),
dataframe.to_csv("fifa_players.csv")
```

```python
def addPlayersFromPage(page_json, dataframe):
    for player in page_json["items"]:
        dataframe.loc[len(dataframe)] = [player["id"], player["firstName"], player["lastName"], player["rating"], player["position"], player["composure"], player["acceleration"], player["aggression"], player["agility"], player["balance"], player["ballcontrol"], player["crossing"], player["curve"], player["dribbling"], player["finishing"], player["freekickaccuracy"], player["jumping"], player["longpassing"], player["longshots"], player["marking"], player["penalties"], player["positioning"], player["potential"], player["reactions"], player["shortpassing"], player["shotpower"], player["slidingtackle"], player["sprintspeed"], player["standingtackle"], player["stamina"], player["strength"], player["vision"], player["volleys"]]

def dataframeColumns():
    return ["Fifa18id","Forename", "LastName", "Rating", "Position", "Composure", "Acceleration", "Aggression", "Agility", "Balance", "BallControl", "Crossing", "Curve", "Dribbling", "Finishing", "FreekickAccuracy", "Jumping", "Longpassing", "Longshots", "Marking", "Penalties", "Positioning", "Potential", "Reactions", "Shortpassing", "Shotpower", "Slidingtackle", "Sprintspeed", "Standingtackle", "Stamina", "Strength", "Vision", "Volleys"]
```
